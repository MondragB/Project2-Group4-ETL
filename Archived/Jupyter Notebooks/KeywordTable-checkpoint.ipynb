```python
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine
```

```python
fp = "keywords.csv"
df = pd.read_csv(fp, delimiter=',')
df.drop(df.index[df['keywords'] == '[]'], inplace=True)
df
```

```python
listofrows = [list(row) for row in df.values]
print(listofrows[0])
```

```python
lod = eval(listofrows[0][1])
lod
```

```python
lod2 = lod[0]
lod2 
```

```python
df2 = pd.DataFrame(lod)
df2
```

```python
list_of_genre_dataframes = []
for row in listofrows:
    keyword_id = eval(row[1])
    record_df = pd.DataFrame(keyword_id)
    record_df['movie_id'] = row[0]
    list_of_genre_dataframes.append(record_df)

keyword_table = pd.concat(list_of_genre_dataframes, sort=False)
keyword_table.reset_index(inplace=True, drop=True)
keyword_table.head(10)
```

```python
keyword_table2 = keyword_table.rename(columns={'id':'keyword_id','name':'word'})
keyword_table2
```

```python
keywords_table = keyword_table2.drop(columns=['movie_id'])
keywords_table
```

```python
keyword_movie_junction_table = keyword_table2.drop(columns=['word'])
keyword_movie_junction_table
```

```python
engine = create_engine("postgresql://postgres:ss2567@localhost:5432/keywords_db")
engine.table_names()
```

```python
keywords_table.to_sql(name='keywords', con=engine, if_exists='replace', index=False)
keyword_movie_junction_table.to_sql(name='keyword_movie_junction', con=engine, if_exists='replace', index=False)
```
